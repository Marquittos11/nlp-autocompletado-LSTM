```python
# --- Fase 1: Preprocesamiento del Texto ---
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
import pickle
import PyPDF2
import re

# 1. ConfiguraciГіn de rutas y extracciГіn
# Ajustado para la carpeta del repositorio clonado
ruta_archivo = "/content/nlp-autocompletado-lstm/documento.pdf"

def extraer_texto_pdf(ruta):
    texto_completo = ""
    with open(ruta, 'rb') as archivo:
        lector_pdf = PyPDF2.PdfReader(archivo)
        for pagina in lector_pdf.pages:
            if pagina.extract_text():
                texto_completo += pagina.extract_text() + " "
    return texto_completo

def limpiar_texto(texto_crudo):
    # NormalizaciГіn: minГәsculas y eliminaciГіn de ruido gramatical [cite: 160-161, 209]
    texto = texto_crudo.lower()
    texto = re.sub(r'[^a-zГЎГ©ГӯГіГәГұ\s]', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

# EjecuciГіn de limpieza
texto_crudo = extraer_texto_pdf(ruta_archivo)
texto = limpiar_texto(texto_crudo)

# 2. TokenizaciГіn y Vocabulario
tokenizer = Tokenizer()
tokenizer.fit_on_texts([texto])
vocab_size = len(tokenizer.word_index) + 1
print(f"TamaГұo del vocabulario: {vocab_size} palabras Гәnicas.")

# 3. GeneraciГіn de Secuencias de Ventana Fija
secuencia_completa = tokenizer.texts_to_sequences([texto])[0]
input_sequences = []
LONGITUD_VENTANA = 6 # 5 palabras de contexto + 1 a predecir

for i in range(LONGITUD_VENTANA, len(secuencia_completa)):
    # Extraemos bloques continuos para modelar la dependencia temporal [cite: 54-55, 212]
    n_gram_sequence = secuencia_completa[i - LONGITUD_VENTANA : i]
    input_sequences.append(n_gram_sequence)

# 4. Padding y EstructuraciГіn de Tensores
# Forzamos maxlen a LONGITUD_VENTANA para asegurar compatibilidad
input_sequences = np.array(pad_sequences(input_sequences, maxlen=LONGITUD_VENTANA, padding='pre'))

# 5. SeparaciГіn de X e y (One-Hot Encoding)
# X: Contexto (5 palabras) | y: Objetivo (Siguiente palabra)
X, y = input_sequences[:, :-1], input_sequences[:, -1]
# Convertimos a categorГӯas para la capa Softmax final [cite: 147, 229-230]
y = tf.keras.utils.to_categorical(y, num_classes=vocab_size)

# 6. Guardar Tokenizer para la AplicaciГіn
with open('tokenizer.pkl', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

print("ВЎProcesamiento completado!")
print(f"Forma de X (Entradas): {X.shape}") # Resultado esperado: (N, 5)
print(f"Forma de Y (Etiquetas): {y.shape}") # Resultado esperado: (N, vocab_size)
```

```python
# --- SEGUNDA FASE: LSTM y Entrenamiento ---
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.optimizers import Adam

# 1. DEFINIR LA ARQUITECTURA
modelo = Sequential()

# Capa 1: Embedding (IncrustaciГіn)
# Convierte identificadores numГ©ricos en vectores densos que capturan la semГЎntica[cite: 183, 205].
# NUEVA MODIFICACIГ“N: Cambiamos max_sequence_len-1 por 5 para que coincida con nuestra ventana fija.
modelo.add(Embedding(input_dim=vocab_size, output_dim=10, input_length=5))

# Capa 2: LSTM (Memoria a Corto y Largo Plazo)
# Orquesta los mecanismos de compuertas para gestionar la memoria secuencial[cite: 184].
# Esta celda utiliza el Carrusel de Error Constante (CEC) para evitar el desvanecimiento del gradiente [cite: 95-96].
modelo.add(LSTM(units=50))

# Capa 3: Capa Densa (Salida)
# Proyecta los datos contra el lГ©xico total usando Softmax para obtener probabilidades [cite: 147, 187, 227-230].
modelo.add(Dense(units=vocab_size, activation='softmax'))

# 2. COMPILAR EL MODELO
# Usamos categorical_crossentropy porque nuestras etiquetas 'y' estГЎn en formato One-Hot[cite: 179].
modelo.compile(loss='categorical_crossentropy',
               optimizer=Adam(learning_rate=0.01),
               metrics=['accuracy'])

# Imprimimos un resumen de la arquitectura
print("--- RESUMEN DE LA ARQUITECTURA ---")
modelo.summary()

# 3. ENTRENAR EL MODELO
# La red itera sobre los datos (epochs) para minimizar el error y aprender patrones humanos[cite: 143, 179].
print("\nIniciando el entrenamiento (esto puede tomar unos segundos)...")
historial = modelo.fit(X, y, epochs=500, verbose=1)
print("ВЎEntrenamiento finalizado con Г©xito!")
# --- NUEVA MODIFICACIГ“N PARA GENERAR EL CEREBRO ---
modelo.save('modelo_autocompletado.h5')
print("рҹ§  ВЎEl cerebro del modelo ha sido guardado como 'modelo_autocompletado.h5'!")
```

```python
# --- TERCERA FASE: Generador de Autocompletado ---
import numpy as np

def predecir_coherente(texto_semilla, n_palabras=4, temperatura=0.6):
    for _ in range(n_palabras):
        tokens = tokenizer.texts_to_sequences([texto_semilla])[0]
        tokens = pad_sequences([tokens], maxlen=5, padding='pre')

        # Obtenemos las probabilidades de la Softmax [cite: 147, 230]
        probas = modelo.predict(tokens, verbose=0)[0]

        # Aplicamos temperatura para evitar el "la la la" [cite: 241-242, 261]
        probas = np.log(probas + 1e-7) / temperatura
        exp_probas = np.exp(probas)
        probas = exp_probas / np.sum(exp_probas)

        # Elegimos una palabra basada en la nueva distribuciГіn
        indice = np.random.choice(range(vocab_size), p=probas)

        palabra_final = ""
        for palabra, i in tokenizer.word_index.items():
            if i == indice:
                palabra_final = palabra
                break
        texto_semilla += " " + palabra_final
    return texto_semilla

# --- PRUEBA EN VIVO ---
# Estas pruebas demuestran la capacidad de la LSTM para capturar patrones secuenciales [cite: 142-143].
print(predecir_coherente("Francisco CantuГұa fue"))
```
